```python
# Traduzir as categorias da coluna 'senioridade'
traducoes_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(traducoes_senioridade)
display(df['senioridade'].value_counts())
```

```python
# Traduzir as categorias da coluna 'tipo_contrato'
traducoes_tipo_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['tipo_contrato'] = df['tipo_contrato'].replace(traducoes_tipo_contrato)
display(df['tipo_contrato'].value_counts())
```

```python
# Traduzir as categorias da coluna 'tamanho_empresa'
traducoes_tamanho_empresa = {
    'M': 'Media',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducoes_tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```python
# Traduzir as categorias da coluna 'remoto'
traducoes_remoto = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hibrido'
}

df['remoto'] = df['remoto'].replace(traducoes_remoto)
display(df['remoto'].value_counts())
```

```python
#colocando novas infos/ excluindo de acordo com alguma regra
#criando um novo df para exemplo
#importando uma bliblioteca ´para manipulação numérica
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})
#atentar para o camel case em DataFrame

#preencher valores nulos com a media dos outros
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#round2 = arredondar 2 casas

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
#usando a mediana ao inves de media

df_salarios #mostrar novamente a tabela
```

```python
# Outro exemplo - monitorando temperaturas
df_temperatura = pd.DataFrame({
    'dia': ['SEG', 'TER', 'QUA', 'QUI', 'SEX'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperatura['preenchido_ffill'] = df_temperatura['temperatura'].ffill()
#ffill (foward fill) preenche com o valor anterior

df_temperatura
```

```python
df_temperatura = pd.DataFrame({
    'dia': ['SEG', 'TER', 'QUA', 'QUI', 'SEX'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperatura['preenchido_bfill'] = df_temperatura['temperatura'].bfill()
#bfill (back fill) preenche com o valor posterior

df_temperatura
```

```python
# preenchendo com condição
df_cidades = pd.DataFrame({
  'nome': ["Ana", "Bruno", "Carlos", "Daniele", 'Val'],
  'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, "Belém", ]
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna("Não informada")
df_cidades

#muitas vezes a IA consgue ajudar com algum erro e até corrigir
```

```python
# importanto uma biclioteca para usar dashboards interativos
import plotly.express as px
```

```python
Feito com IA

# Calcular a média salarial por senioridade
media_salario_por_senioridade = df_limpa.groupby('senioridade')['usd'].mean().reset_index()
# reset index pega somente os nokes da coluna

# Ordenar os dados pela média salarial
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']
media_salario_por_senioridade['senioridade'] = pd.Categorical(media_salario_por_senioridade['senioridade'], categories=ordem_senioridade, ordered=True)
media_salario_por_senioridade = media_salario_por_senioridade.sort_values('senioridade')


# Criar o gráfico de barras interativo com Plotly
fig = px.bar(media_salario_por_senioridade,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual (USD)'},
             color='senioridade', # Color bars by seniority level
             category_orders={'senioridade': ordem_senioridade}) # Ensure the correct order

fig.show()
```

```python
# mesmo grafico anterior porem com uma codigo mais simples:

senioridade_media_salario = df_limpa.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial (USD)'}
             )

fig.show()
```

```python
# criando graficos de pizza /torta (geralmente usando com ate 3 categorias)

remoto_contagem = df_limpa['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho',
            )
fig.show()
```

```python
# convertendo em grafico de rosca e colocanco os rotulos
remoto_contagem = df_limpa['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho',
             hole=0.5
            )
fig.update_traces(textinfo='percent+label')
fig.show()
```
